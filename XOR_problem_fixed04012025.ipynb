```python
# We will use two input neurons, one hidden layer, and one output.
# input data
input_data = np.array([[0,0], [0,1], [1,0], [1,1]])
# output data, XOR
output_data = np.array([[0], [1], [1], [0]])

# Checking Shapes, got a value error first time
print("input_data shape:", input_data.shape) # We expect(4,2)
print("output_data shape:", output_data.shape) # Expect (4,1)

# A Sequential Model
model = Sequential()
model.add(Dense(8, input_dim = 2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

# We will now compile the model.
optimizer = Adam(learning_rate = 0.01)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Let's train this model.
model.fit(input_data, output_data, epochs=1000, verbose=2)

# We will now evaluate this model.
loss, accuracy = model.evaluate(input_data, output_data)

print(f"Model Accuracy: {accuracy*100:.2f}%")

# Making Predictions for XOR
predictions = model.predict(input_data)
print("Predictions are: ")
print(np.round(predictions)) # We are rounding to ensure our results are binary.
```
